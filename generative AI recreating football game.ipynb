## Gait Length Analysis for Labels :

```python
import numpy as np
import plotly.express as px
import pandas as pd

# Assuming you have already collected 'norm' values into donnees1
# Initialize acceptable gait length range in seconds
min_gait_length = 0.1  # Minimum acceptable gait length (e.g., 0.1 seconds)
max_gait_length = 3.0  # Maximum acceptable gait length (e.g., 3.0 seconds)

# Lists to store gait lengths, labels, and label statuses
gait_lengths = []
labels = []
label_statuses = []

# Loop through the data to check gait length for each label
for echantillon in donnees1:
    selected_label = echantillon['label']
    norm_data = echantillon['norm']  # Use the 'norm' data for the selected label

    # Assuming a constant sampling rate of 50Hz
    sampling_rate = 50  # Hz

    # Determine the dynamic threshold (10% of the maximum accelerometer value)
    dynamic_threshold = 0.1 * max(norm_data)

    # Initialize lists to store indices of detected peaks and valleys
    peaks = []
    valleys = []

    # Detect peaks and valleys in the accelerometer data
    for i in range(1, len(norm_data) - 1):
        if (
            norm_data[i] > norm_data[i - 1]
            and norm_data[i] > norm_data[i + 1]
            and norm_data[i] > dynamic_threshold
        ):
            peaks.append(i)
        elif (
            norm_data[i] < norm_data[i - 1]
            and norm_data[i] < norm_data[i + 1]
        ):
            valleys.append(i)

    # Calculate time durations between consecutive peaks (stance phases) and valleys (swing phases)
    durations_between_peaks = [(i - j) / sampling_rate for i, j in zip(peaks[1:], peaks[:-1])]
    durations_between_valleys = [(i - j) / sampling_rate for i, j in zip(valleys[1:], valleys[:-1])]

    # Calculate the mean time duration for both the stance phase (between peaks) and swing phase (between valleys)
    if len(durations_between_peaks) > 0:
        mean_stance_phase_duration = sum(durations_between_peaks) / len(durations_between_peaks)
    else:
        mean_stance_phase_duration = 0.0  # Set a default value if no peaks are detected

    if len(durations_between_valleys) > 0:
        mean_swing_phase_duration = sum(durations_between_valleys) / len(durations_between_valleys)
    else:
        mean_swing_phase_duration = 0.0  # Set a default value if no valleys are detected

    # Calculate the gait length (sum of stance phase and swing phase)
    gait_length = mean_stance_phase_duration + mean_swing_phase_duration

    # Append gait length, label, and label status to the respective lists
    gait_lengths.append(gait_length)
    labels.append(selected_label)
    
    if min_gait_length <= gait_length <= max_gait_length:
        label_status = 'Within Range'
        within_range = True
    else:
        label_status = 'Outside Range'
        within_range = False
    
    label_statuses.append(label_status)

    # Add the calculated information to the original echantillon dictionary
    
    echantillon['within_range'] = within_range  # Add 'within_range' field
    echantillon['gait_length'] = gait_length


# Create a DataFrame for visualization
df = pd.DataFrame({'Label': labels, 'Gait Length': gait_lengths, 'Label Status': label_statuses})

# Create a bar chart using Plotly Express
fig = px.bar(df, x='Label', y='Gait Length', color='Label Status', title='Gait Lengths for All Labels')

# Remove the values of 'y' from the graph
fig.update_yaxes(title_text='', showticklabels=False)

# Show the bar chart
fig.show()





```

## Suspicious Action Sequence Detection :

```python
# Initialize variables
last_action = None
sequence_count = 0
suspicious_sequences = []
threshold = 5

# Create a new list to store modified data with flags
donnees1_modified = []

# Iterate through the data
for data_point in donnees1:
    label = data_point['label']

    if label == last_action:
        sequence_count += 1
        if sequence_count >= threshold:
            # Flag the current data point as suspicious
            data_point['is_suspicious'] = True
    else:
        if sequence_count >= threshold:
            suspicious_sequences.append((last_action, sequence_count))
        last_action = label
        sequence_count = 1

    # Append the modified data point to donnees1_modified
    donnees1_modified.append(data_point)

# Check for the last sequence
if sequence_count >= threshold:
    suspicious_sequences.append((last_action, sequence_count))
    # Flag the last data point in the sequence as suspicious
    donnees1_modified[-1]['is_suspicious'] = True

# Print the suspicious sequences
for sequence in suspicious_sequences:
    print(f"Suspicious sequence: {sequence[0]} repeated {sequence[1]} times successively.")




```

## Random Forest Classifier for Action Recognition :

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Define a function for data augmentation
def augment_data(sample):
    augmented_samples = []
    for _ in range(10):  # Augment with 10 new samples (adjust as needed)
        augmented_sample = np.array(sample['norm'])
        
        # Apply random shifts to the time series data
        shift = np.random.randint(-5, 5)  # Adjust the shift range
        augmented_sample = np.roll(augmented_sample, shift)
        
        # You can add more transformations here, such as scaling, noise, etc.
        
        augmented_samples.append({'label': sample['label'], 'norm': augmented_sample})
    
    return augmented_samples

# Augment the data
augmented_data = []
for sample in donnees1:
    augmented_samples = augment_data(sample)
    augmented_data.extend(augmented_samples)

# Determine the maximum length of the augmented samples
max_length = max(len(sample['norm']) for sample in augmented_data)

# Pad all augmented samples to have the same length as the longest one
X = np.array([np.pad(sample['norm'], (0, max_length - len(sample['norm'])), 'constant') for sample in augmented_data])
y = np.array([sample['label'] for sample in augmented_data])

# Encode labels to numerical values
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.30, random_state=42)

# Initialize and train a Random Forest model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate precision, recall, and F1-score for each class
class_names = label_encoder.classes_
report = classification_report(y_test, y_pred, target_names=class_names)

print("Classification Report:\n", report)


```
